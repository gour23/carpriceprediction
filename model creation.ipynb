```python
def corelation(dataset,thershold):
    col_cor = set()   # set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i,j])>thershold:      # both negative and positive 
                colname = corr_matrix.columns[i]       # getting the name of columns
                col_cor.add(colname)
    return col_cor

    
```

```python

from sklearn.model_selection import RandomizedSearchCV
```

```python
# HYPERPARAMETERS

# number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start=100,stop=1200,num=12)]

# max number of levels in tree
max_depth = [int(x) for x in np.linspace(5,30,num=6)]
# max_depth.append(None)

#min number of samples required to split a node
min_sampls_split = [2,5,10,15,100 ]

#min number of samples required at each leaf node
min_sampls_leaf = [1,2,5,10 ]
```

```python
# Create the Random Grid

random_grid = {
    "n_estimators":n_estimators,
    "max_depth":max_depth,
    "min_samples_split":min_sampls_split,
    "min_samples_leaf":min_sampls_leaf
} 
print(random_grid)
```

```python
rf_model = RandomizedSearchCV(estimator=rf , param_distributions=random_grid ,scoring="neg_mean_squared_error" , n_iter=10 , cv=5,verbose=2,random_state=42,n_jobs=1)
```

```python
rf_model.fit(X_train,y_train)
```
